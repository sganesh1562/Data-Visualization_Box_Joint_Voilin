```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
df=pd.read_csv("/content/tips.csv")
```

```python
snr.jointplot(x='total_bill',y='tip',data=df)
plt.show()
```
